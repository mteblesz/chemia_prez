```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dane = np.genfromtxt('13.csv', delimiter=';')
x = dane[:, 0] # pierwsza kolumna
y = dane[:, 1] # druga kolumna
plt.scatter(x, y) # wykres punktowy

#modyfikowanie wykresu

# 1) Ustawienie skal logarytmicznych
plt.xscale('log')
plt.yscale('log')

plt.xlabel('X (log scale)')
plt.ylabel('Y (log scale)')
plt.title('Wykres logarytmiczny')

# 2) Zmiana kolorów pnktów
plt.scatter(x, y, color='red')

# 3) Dodanie siatki
plt.grid(True)
plt.show()
```
